```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

from sktime.performance_metrics.forecasting import (
    mean_absolute_percentage_error,
    mean_absolute_error,
)

import seaborn as sns
import matplotlib.pyplot as plt
import random
from operator import itemgetter 
```

```python
class CustRNN(nn.Module):
    def __init__(self, input_shape, output_shape, n_dense_1, n_dense_2, rnn_hidden_size):
        super(CustRNN, self).__init__()
        self.rnn_layer = nn.RNN(input_size=input_shape,hidden_size=rnn_hidden_size,num_layers=3, nonlinearity='relu')
        self.layer_1 =  nn.Linear(rnn_hidden_size, n_dense_1)
        self.bn1 = nn.BatchNorm1d(num_features=n_dense_1)
        self.layer_2 =  nn.Linear(n_dense_1, n_dense_2)
        self.layer_3 =  nn.Linear(n_dense_2, output_shape, dtype= torch.float32)
        
        
    def forward(self, input, hidden_prev):
        x, hidden_nxt = self.rnn_layer(input,hidden_prev)
        # print(x.shape)
        x = F.relu(self.layer_1(x))
        # if x.dim()>1:
        #     x = self.bn1(x)
        x = F.relu(self.layer_2(x))
        outputs = self.layer_3(x)
        return outputs, hidden_nxt
```

```python
# Device
DEVICE = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")
```

```python
training_data = pd.read_csv('../train_df_2.csv')
training_data['arrival_date'] = pd.to_datetime(training_data['arrival_date'])
training_data = training_data.drop(['Unnamed: 0'], axis=1).sort_values('arrival_date')


PREDICTION_HORIZON = 30

TARGET_COLS = [f'RV{i}' for i in range(PREDICTION_HORIZON)] # ['RV0']
FEATURE_COLS = [f'RV{i}' for i in range(PREDICTION_HORIZON,100 + 1)]
target_data = training_data[TARGET_COLS]
feature_data = training_data[FEATURE_COLS]
```

```python
target_data_tensor = torch.tensor(target_data.to_numpy(), dtype= torch.float32)
feature_data_tensor = torch.tensor(feature_data.to_numpy(), dtype= torch.float32)
```

### Training Loop

```python
# training using the entire data set in each epoch
HIDDEN_SIZE = PREDICTION_HORIZON
rnn_model = CustRNN( input_shape=len(FEATURE_COLS), output_shape=len(TARGET_COLS), n_dense_1=100, n_dense_2=100, rnn_hidden_size=HIDDEN_SIZE)
rnn_model = rnn_model.to(DEVICE)
target_data_tensor = target_data_tensor.to(DEVICE)
feature_data_tensor = feature_data_tensor.to(DEVICE)


optimizer = torch.optim.Adam(rnn_model.parameters(), lr = 0.001)

hidden_prev = torch.randn(3, HIDDEN_SIZE)
hidden_prev = hidden_prev.to(DEVICE)

for epoch in range(300):
    pred,hidden_prev = rnn_model(feature_data_tensor,hidden_prev)
    hidden_prev = hidden_prev.detach()
    
    cost = nn.functional.mse_loss(pred,target_data_tensor)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100==0:
        print('epoch:',epoch,' cost:',cost)

```

#### Checking for overfitting

```python
# checking training SMAPE for overfitting
insample_pred,hidden_prev = rnn_model(feature_data_tensor,hidden_prev)

insample_comparison = training_data.copy()
insample_comparison['pred'] = insample_pred.cpu().detach().numpy()[:,0]

print('in sample SMAPE:',mean_absolute_percentage_error(y_pred=insample_comparison['pred'], y_true=insample_comparison[TARGET_COLS[0]], symmetric=True))
print('in sample MAE:',mean_absolute_error(y_pred=insample_comparison['pred'], y_true=insample_comparison[TARGET_COLS[0]]))


plt.figure(figsize=(30,8))
sns.lineplot(data=insample_comparison, x = 'arrival_date', y=TARGET_COLS[0], label='True value')
sns.lineplot(data=insample_comparison, x = 'arrival_date', y='pred', label='Pred value')
plt.title('In sample Predictions')
plt.legend()
```

#### Testing the RNN on out of sample predictions

```python
test_df = pd.read_csv('../test_df_2.csv').reset_index(drop=True).drop('Unnamed: 0',axis=1)
test_df['arrival_date'] = pd.to_datetime(test_df['arrival_date'])
```

```python
test_data_tensor = torch.tensor(test_df[FEATURE_COLS].to_numpy(), dtype=torch.float32)
test_data_tensor = test_data_tensor.to(DEVICE)

preds, hidden_output = rnn_model(test_data_tensor,  hidden_prev)
preds = preds.cpu().detach().numpy()
pred_df = test_df.copy()
pred_df['pred'] = preds[:,0]


print('SMAPE:',mean_absolute_percentage_error(y_pred=pred_df['pred'], y_true=pred_df[TARGET_COLS[0]], symmetric=True))
print('MAE:',mean_absolute_error(y_pred=pred_df['pred'], y_true=pred_df[TARGET_COLS[0]]))

plt.figure(figsize=(30,8))
sns.lineplot(data=pred_df, x = 'arrival_date', y=TARGET_COLS[0], label='True value')
sns.lineplot(data=pred_df, x = 'arrival_date', y='pred', label='Pred value')
plt.title('Out of sample Predictions')
plt.legend()
```

```python
# running a realistic sliding window loop, 
# simulating a inferencing done every X days(X is the parameter freq in pd.date_range )

test_window_starts = pd.date_range(start=test_df.arrival_date.min(), 
                                   end=test_df.arrival_date.max(), 
                                   freq='5D')

all_eval_dfs = []
for window_start in test_window_starts:
    window_end = window_start+pd.to_timedelta(PREDICTION_HORIZON,unit='days')
    print(f"inferencing on {window_start} | target_stay_date: {window_end}")
    test_row = test_df[test_df.arrival_date==window_end]

    if len(test_row)>0:

        slide_test_data_tensor = torch.tensor(test_row[FEATURE_COLS].to_numpy(), dtype=torch.float32)
        slide_test_data_tensor = slide_test_data_tensor.to(DEVICE)

        slide_pred, hidden_output = rnn_model(slide_test_data_tensor,  hidden_prev)

        eval_df = test_row[TARGET_COLS].melt().rename(columns={'variable':'target', 'value':'y_true'})
        eval_df['y_pred'] = slide_pred.cpu().detach().numpy()[0]
        eval_df['eval_start'] = window_start
        eval_df['arrival_date'] = window_end
        all_eval_dfs.append(eval_df)

slide_output_df = pd.concat(all_eval_dfs)


    
```
