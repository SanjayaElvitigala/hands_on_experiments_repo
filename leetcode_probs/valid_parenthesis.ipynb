```python
sample_test_cases = [
    ("()", True),
    ("()[]{}", True),
    ("(]",False),
    ("{[]}", True),
    ("([)]", False),
    ("]",False),
    ("({{{{}}}))", False),
    ("]()", False),
    ("[])", False),
    ("[([]])", False),
    ("(}{)", False),
    ("(([]){})",True),
    ("(){}}{",False),
]
```

```python
def isValid_v3( s: str) -> bool:
    open_brac_chars = {'(': ')', '{' : '}', '[' :']'}
    close_brac_chars = { ')':'(', '}':'{', ']':'['}

    all_chars = open_brac_chars.copy()
    all_chars.update(close_brac_chars)

    input_chars = list(s)
    full_len = len(input_chars)

    uniq_chars,char_cnts = np.unique(input_chars, return_counts=True)
    char_cnts = dict(zip(uniq_chars,char_cnts))


    open_brac_stack = []
    
    for char in char_cnts:

        if char in open_brac_chars.keys():
            opp_char = open_brac_chars[char]
        else:
            opp_char = close_brac_chars[char]
        if char_cnts[char]!=char_cnts.get(opp_char):
            print('exe 1')
            return False

    for char in input_chars:

        if char in open_brac_chars.keys():
            open_brac_stack.append(char)
        else:
            try:
                if open_brac_stack[-1]!=close_brac_chars[char]:
                    return False
                else:
                    open_brac_stack.pop(-1)
            except Exception as e:
                print(e)
                return False
                
    
    return True
        
```

```python
for test_case, expected_val in sample_test_cases:
    print('______')
    print(test_case,expected_val)
    print('assertion output: ',expected_val==isValid_v3(test_case))
```
